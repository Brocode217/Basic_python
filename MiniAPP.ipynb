```python
# Initialize data structures
customers = {}
balances = {}
transaction_types = {"deposit", "withdrawal"}
transaction_history = {}

# Function to add a new customer
def add_customer():
    account_number = int(input("Enter Customer ID (Account Number): "))
    if account_number in customers:
        print("Account number already exists.")
        return
    name = input("Enter Name: ")
    account_type = input("Enter Account Type (e.g., Savings/Checking): ")
    initial_balance = float(input("Enter Initial Balance: "))
    customers[account_number] = (name, account_type)
    balances[account_number] = initial_balance
    transaction_history[account_number] = []
    print(f"Customer {name} added with account number {account_number}.")

# Function to deposit money
def deposit():
    account_number = int(input("Enter Account Number for Deposit: "))
    if account_number in balances:
        amount = float(input("Enter Amount to Deposit: "))
        date = input("Enter Date of Deposit (YYYY-MM-DD): ")
        balances[account_number] += amount
        transaction_history[account_number].append({
            "type": "deposit",
            "amount": amount,
            "date": date
        })
        transaction_types.add("deposit")
        print(f"Deposited {amount} to account {account_number}. New balance: {balances[account_number]}")
    else:
        print("Account not found.")

# Function to withdraw money
def withdraw():
    account_number = int(input("Enter Account Number for Withdrawal: "))
    if account_number in balances:
        amount = float(input("Enter Amount to Withdraw: "))
        if balances[account_number] >= amount:
            date = input("Enter Date of Withdrawal (YYYY-MM-DD): ")
            balances[account_number] -= amount
            transaction_history[account_number].append({
                "type": "withdrawal",
                "amount": amount,
                "date": date
            })
            transaction_types.add("withdrawal")
            print(f"Withdrew {amount} from account {account_number}. New balance: {balances[account_number]}")
        else:
            print("Insufficient balance.")
    else:
        print("Account not found.")

# Function to view balance
def view_balance():
    account_number = int(input("Enter Account Number to View Balance: "))
    if account_number in balances:
        print(f"Account {account_number} balance: {balances[account_number]}")
    else:
        print("Account not found.")

# Function to view transaction history
def view_transaction_history():
    account_number = int(input("Enter Account Number to View Transaction History: "))
    if account_number in transaction_history:
        print(f"Transaction history for account {account_number}:")
        for transaction in transaction_history[account_number]:
            print(transaction)
    else:
        print("Account not found.")

# Main menu to perform actions
def main():
    while True:
        print("\n--- Banking Application ---")
        print("1. Add New Customer")
        print("2. Deposit Money")
        print("3. Withdraw Money")
        print("4. View Balance")
        print("5. View Transaction History")
        print("6. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            add_customer()
        elif choice == "2":
            deposit()
        elif choice == "3":
            withdraw()
        elif choice == "4":
            view_balance()
        elif choice == "5":
            view_transaction_history()
        elif choice == "6":
            print("Exiting the application.")
            break
        else:
            print("Invalid choice. Please try again.")

# Run the banking application
main()
```
